```python
first_class = test_df[test_df['Pclass'] == 1]
second_class = test_df[test_df['Pclass'] == 2]
third_class = test_df[test_df['Pclass'] == 3]
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
y = [len(first_class), len(second_class), len(third_class)]
l = ['Passenger Class 1 ', 'Passenger Class 2', 'Passenger Class 3']
ax.pie(y, labels = l, autopct='%1.2f%%')
plt.show()
```

# My Analysis
### Passenger Class
From the data it is observed that there were more buyers in Queenstown then other for the ship's tickets.
There were also more people in passenger class 3 because the fares were cheaper.
The data clearly shows the class system aboard the Titanic, with significant differences in fares and accommodations between first, second, and third-class passengers. 
### Age range
The passenger list includes a wide range of ages, from infants to elderly, and a mix of genders. This suggests that the Titanic was carrying a diverse group of travelers, possibly including families, solo travelers, and group
### Fare prices
The vast difference in fares (from as low as 3.17 to as high as 512.33) indicates a significant economic gap among passengers. This aligns with the class system and the luxury status of the Titanic.
